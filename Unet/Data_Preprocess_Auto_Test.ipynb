```python
import glob
import cv2
import matplotlib.pyplot as plt
import numpy as np
#import scipy
from scipy import misc
%matplotlib inline
```

```python
DATA_ROOT = "/Users/cesare.chung/int/01_CODE_UNET/NEW/"
#
# Testing - preprocessing
#
imgs_Testing  = sorted(glob.glob(DATA_ROOT+"/testData_tftf/resize2/*_test_resized_test.tif"))
#
# Import data
print('Loading testing dataset...\n')
#
X_test=[]
for index,entry in enumerate(imgs_Testing):
    img = misc.imread(entry)
    X_test.append(img)

#
X_test=np.array(X_test)
#
plt.imshow(X_test[19,...,2], aspect="auto")   
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler

stdsc = StandardScaler()

X_test_std=[]
for i in range(0,20):
    img_std = stdsc.fit_transform(X_test[i,...,0])
    X_test_std.append(img_std)

#
X_test_std=np.array(X_test_std)
#
```

```python
print("TEST  data  shape:", X_test.shape)
```

```python
print("TEST data  shape (Scaled):", X_test_std.shape)
```

```python
for i in range(6,10):
    plt.imshow(X_test[i,...,0], aspect="auto")
    plt.title("Un-scaled")
    plt.show()
    plt.imshow(X_test_std[i,...,], aspect="auto")
    plt.title("Scaled")
    plt.show()
```
