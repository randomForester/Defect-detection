```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap
#import plotly.plotly as py  # tools to communicate with Plotly's server
```

```python
df = pd.read_csv('/Users/cesare.chung/int/01_CODE_UNET/TensorBoard_unet/accuracy200.csv', header=None)
```

```python
df.head()
```

```python
y = df.iloc[1:, 2].values
X = df.iloc[1:, 1].values
'''
y = df.iloc[100:, 2].values
X = df.iloc[100:, 1].values
'''
#
#print(y)
#
# 'o' 'rs' 'g^' 'bo' 'k' '-' 'r-' 'r--'
#
plt.title('Accuracy - Epoch Training Phase')
plt.plot(X,y, '-', label='Training Phase (Epoch = 200)')

plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend()
plt.tight_layout()
plt.grid(True)
plt.savefig('/Users/cesare.chung/Desktop/_Accuracy_unet.png', dpi=300)
plt.show()
```
